```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data= pd.read_csv("banking_data.csv")
data.head()
```

### 1. Distribution of age among clients

```python
sns.set(style="whitegrid")

# 1. Distribution of age among clients
plt.figure(figsize=(10, 5))
sns.histplot(data['age'], bins=30, kde=True, color='blue')
plt.title("Distribution of Age Among Clients")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

the distribution is right-skewed


```python
# Get the minimum and maximum age values
age_min = data['age'].min()
age_max = data['age'].max()

# Get the most common age range (Interquartile Range - IQR)
age_q1 = data['age'].quantile(0.25)
age_q3 = data['age'].quantile(0.75)

(age_min, age_max, age_q1, age_q3)
```

25% data of age criteria falls below 33 and 75% values fall below 48. with minimum and maximum value of 18 and 95

## 2. Job type distribution

```python
# 2. Job type distribution
plt.figure(figsize=(12, 6))
sns.countplot(y=data['job'], order=data['job'].value_counts().index, palette="viridis")
plt.title("Job Type Distribution Among Clients")
plt.xlabel("Count")
plt.ylabel("Job Type")
plt.show()
```

The job distribution shows that certain professions, such as "blue-collar" and "management," have a higher representation among clients.

#Boxplot for Balance:

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.boxplot(data['balance'])
plt.title('Distribution of Average Yearly Balance')
plt.xlabel('Balance (Euros)')
plt.show()
```

#Pie Charts for Loan Types:

```python
housing_counts = data['housing'].value_counts()
loan_counts = data['loan'].value_counts()

# Housing Loan Pie Chart
plt.pie(housing_counts, labels=housing_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Housing Loan Distribution')
plt.show()

# Personal Loan Pie Chart
plt.pie(loan_counts, labels=loan_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Personal Loan Distribution')
plt.show()
```

```python
# Analyzing communication types, last contact day distribution, contact month, last contact duration, and campaign contacts

# 1. Communication types used for contacting clients
communication_types = data['contact'].value_counts()

# 2. Distribution of the last contact day of the month
contact_day_distribution = data['day'].describe()

# 3. Last contact month variation among the clients
contact_month_variation = data['month'].value_counts()

# 4. Distribution of the duration of the last contact
duration_distribution = data['duration'].describe()

# 5. Number of contacts performed during the campaign for each client
campaign_contacts_distribution = data['campaign'].describe()

communication_types, contact_day_distribution, contact_month_variation, duration_distribution, campaign_contacts_distribution
```

Code for Bar Charts (Communication Type and Contact Month)

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Communication Type Distribution
communication_counts = data['contact'].value_counts()
plt.figure(figsize=(8, 6))
sns.barplot(x=communication_counts.index, y=communication_counts.values, palette='viridis')
plt.title('Communication Types Used for Contacting Clients')
plt.xlabel('Communication Type')
plt.ylabel('Number of Clients')
plt.show()

# Last Contact Month Distribution
month_counts = data['month'].value_counts()
plt.figure(figsize=(8, 6))
sns.barplot(x=month_counts.index, y=month_counts.values, palette='coolwarm')
plt.title('Distribution of Last Contact Month')
plt.xlabel('Month')
plt.ylabel('Number of Clients')
plt.xticks(rotation=45)
plt.show()
```

```python
#Distribution of the number of days passed since the client was last contacted
plt.figure(figsize=(10, 6))
sns.histplot(data['pdays'], bins=30, kde=True, color='blue')
plt.title('Distribution of Days Passed Since Last Contact')
plt.xlabel('Days Passed')
plt.ylabel('Frequency')
plt.show()

# Number of contacts performed before the current campaign for each client
plt.figure(figsize=(10, 6))
sns.histplot(data['previous'], bins=30, kde=False, color='green')
plt.title('Number of Contacts Before Current Campaign')
plt.xlabel('Number of Contacts')
plt.ylabel('Frequency')
plt.show()

# Outcomes of the previous marketing campaigns
previous_outcome_counts = data['poutcome'].value_counts()
plt.figure(figsize=(8, 6))
sns.barplot(x=previous_outcome_counts.index, y=previous_outcome_counts.values, palette='Set2')
plt.title('Outcomes of Previous Marketing Campaigns')
plt.xlabel('Outcome')
plt.ylabel('Number of Clients')
plt.show()

# Distribution of clients who subscribed to a term deposit vs. those who did not
subscription_counts = data['y'].value_counts()
plt.figure(figsize=(8, 6))
sns.barplot(x=subscription_counts.index, y=subscription_counts.values, palette='Set3')
plt.title('Subscription to Term Deposit')
plt.xlabel('Subscription')
plt.ylabel('Number of Clients')
plt.show()

# Correlations between attributes and the likelihood of subscribing
# Select only numeric columns for correlation analysis
numeric_data = data.select_dtypes(include=[np.number])

# Compute the correlation matrix
correlation_matrix = numeric_data.corr()

plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, fmt='.2f', cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()

# Examining specific correlations with the target variable
if 'y_binary' in numeric_data.columns:
    target_correlation = correlation_matrix['y_binary'].sort_values(ascending=False)
    print("Correlations with Subscription Likelihood:\n")
    print(target_correlation)
else:
    print("The 'y_binary' column is not in the numeric dataset.")
```
